# Pandas Series

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

# Series Indexing 

```python
# integer indexing
x = pd.Series([12,13,14,35,46,57,58,79,9])
x[0]   # 12
x[1]   # 13
x[2]   # 14
```

x.iloc[-1] returns the last element in the Series x by position. iloc always uses numeric positions, not labels.

So:

x.iloc[0] is the first value,
x.iloc[1] is the second value,
x.iloc[-1] is the last value

```python
# negative indexing
x.iloc[-1]
```
